```python
import numpy as np
from math import factorial
```

```python
def combination(n: int, k: int) -> int:
    """
    Сочетания из n по k
    """
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
def binomial_distribution(n: int, k: int, prob: float) -> float:
    """
    Биноминальное распределение
    """
    return combination(n, k) * (prob ** k) * (1 - prob) ** (n - k)
```

```python
def poisson_distribution(n: int, m: int, prob: float) -> float:
    """
    Распределение Пуассона
    """
    lambda_ = n * prob
    return (lambda_ ** m / np.math.factorial(m)) * np.exp(-lambda_)
```

#### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
shoot_probability = binomial_distribution(100, 85, .8)

print(f"Ответ: {shoot_probability:1.5}")
```

#### 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. 

Какова вероятность, что ни одна из них не перегорит в первый день?

```python
good_lamp_probability = 1 - poisson_distribution(5000, 1, .0004)

print(f"Ответ: {good_lamp_probability:1.5}")
```

Какова вероятность, что перегорят ровно две?

```python
two_bad_lamp_probability = poisson_distribution(5000, 2, .0004)

print(f"Ответ: {two_bad_lamp_probability:1.5}")
```

#### 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. 

Какова вероятность того, что все мячи белые?

```python
first_box_output_two_white = combination(7, 2) / combination(10, 2)

second_box_output_two_white = combination(9, 2) / combination(11, 2)

two_white_probability = first_box_output_two_white * second_box_output_two_white
print(f"\nОтвет: {two_white_probability:1.5}")
```

Какова вероятность того, что ровно два мяча белые?

```python
# вероятность достать 1 белый мяч из первой коробки
one_box_one_white_probability = (combination(7, 1) * combination(3, 1)) / combination(10, 2)
# вероятность достать 2 белых мяча из первой коробки
one_box_two_white_probability = (combination(7, 2) * combination(3, 0)) / combination(10, 2)

# вероятность достать 1 белый мяч из второй коробки
two_box_one_white_probability = (combination(9, 1) * combination(3, 1)) / combination(11, 2)
# вероятность достать 2 белых мяча из второй коробки
two_box_two_white_probability = (combination(9, 2) * combination(3, 0)) / combination(11, 2)

# вероятность достать по одному белому из каждой коробки или сразу два из одной из коробок
two_white_probability = (one_box_one_white_probability * two_box_one_white_probability) * (one_box_two_white_probability + two_box_two_white_probability)

print(f"Ответ: {two_white_probability:1.5}")
```

Какова вероятность того, что хотя бы один мяч белый?

```python
first_box_output_one_white = 1 - (combination(3, 2) / combination(10, 2))

second_box_output_one_white = 1 - (1 / combination(10, 2))

one_white_probability = first_box_output_one_white * second_box_output_one_white
print(f"\nОтвет: {one_white_probability:1.5}")
```
