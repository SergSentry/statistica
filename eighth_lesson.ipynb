```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f, f_oneway
```

#### Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. Даны значения роста в трех группах случайно выбранных спортсменов: Футболисты: 173, 175, 180, 178, 177, 185, 183, 182. Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180. Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
football_players = np.array([173, 175, 180, 178, 177, 185, 183, 182], dtype=np.float64)
fn = len(football_players)
f_mean = football_players.mean()

hockey_players = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180], dtype=np.float64)
hn = len(hockey_players)
h_mean = hockey_players.mean()

weightlifters = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170], dtype=np.float64)
wn = len(weightlifters)
w_mean = weightlifters.mean()

all_data = np.concatenate([football_players, hockey_players, weightlifters])
all_mean = all_data.mean()

n = len(all_data)
k = 3
#

std = np.sum((all_data - all_mean)**2)
print(f"std:\t\t{std:.3f}")

factor_std = ((f_mean - all_mean)**2) * fn + ((h_mean - all_mean)**2) * hn + ((w_mean - all_mean)**2) * wn
print(f"factor_std:\t{factor_std:.3f}")

residual_std = np.sum((football_players - f_mean)**2) + np.sum((hockey_players - h_mean)**2) + np.sum((weightlifters - w_mean)**2)
print(f"residual_std:\t{residual_std:.3f}")

print(f"\nCheck: {factor_std:.3f} + {residual_std:.3f} = {(factor_std + residual_std):.3f} = std:{std:.3f}\n") 

factor_variance = factor_std/(k-1)
print(f"factor_variance:\t{factor_variance:.3f}")

residual_variance = residual_std/(n-k)
print(f"residual_variance:\t{residual_variance:.3f}")

fp = factor_variance / residual_variance
print(f"fp:\t\t\t{fp:.3f}")

#Фишер
print("\nКритерий Фишера")
alpha = .05
df1 = k - 1
df2 = n - k
ft = f.ppf(1 - alpha, df1, df2)
print(f"fp:{fp:.3f}, ft:{ft:.3f}")
if fp > ft:
    print("H1 стат. значима")
else:
    print("H0 стат. не значима")
```
