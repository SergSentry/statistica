```python
from math import factorial
```

```python
def probability(n: int, k: int) -> float:
    """
    Вероятность
    """
    return float(k / n)
```

```python
def combination(n: int, k: int) -> int:
    """
    Сочетания из n по k
    """
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
def arrangements(n: int, k: int) -> int:
    """
    Размещение из n по k
    """
    return int(factorial(n) / factorial(n - k))
```

```python
def permutation(n : int) -> int:
    """
    Перестановка из n
    """
    return int(factorial(n))
```

#### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты.

a) Найти вероятность того, что все карты – крести.

```python
probabality_crosses = combination(13, 4) / combination(52, 4)

print(f"\nВероятность того, что все карты – крести равна: {probabality_crosses:1.5f}")
```

б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз

```python
probabality_ace = 1 - (combination(48, 4) / combination(52, 4))

print(f"\nВероятность, что среди 4-х карт окажется хотя бы один туз равна: {probabality_ace:1.5f}")
```

#### 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

```python
code_variants = combination(10, 3)
probabality_first_enter = probability(code_variants, 1)

print(f"\nВероятность открытия двери с первой попытки равна: {probabality_first_enter:1.5f}")
```

#### 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

```python
probability_color_one_item = probability(15, 9)
probability_color_two_items = probability_color_one_item * probability(14, 8)
probability_color_three_items = probability_color_two_items * probability(13, 7)

print(f"\nВероятность того, что все извлеченные детали окрашены равна: {probability_color_three_items:1.5f}")
```

#### 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
win_probability_one = probability(100, 2)
win_probability_two = win_probability_one * probability(99, 1)

print(f"\nВероятность того, что 2 приобретенных билета окажутся выигрышными равна: {win_probability_two:1.5f}")
```
