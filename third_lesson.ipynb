```python
from math import factorial, sqrt
```

```python
def combination(n: int, k: int) -> int:
    """
    Сочетания из n по k
    """
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
def mean(ar) -> float:
    """
    Среднее арифметическое
    """
    return float(sum(ar)/len(ar))
```

```python
def variance(data, ddof=0) -> float:
    """
    Дисперсия
    """
    m = mean(data)
    n = len(data)
    return float(sum((x - m)**2 for x in data)/(n - ddof))
```

```python
def mean_square_deviation(data, ddof=0) -> float:
    """
    Среднее квадратичное отклонение
    """
    return sqrt(variance(data, ddof))
```

#### 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
test_data = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

test_data_mean = mean(test_data)
print(f"Среднее арифметическое: {test_data_mean:.2f}")

test_data_mean_square_deviation = mean_square_deviation(test_data)
print(f"Среднее квадратичное отклонение: {test_data_mean_square_deviation:.2f}")

test_data_variance = variance(test_data)
print(f"Дисперсия: {test_data_variance:.2f}")

test_data_variance_ddof = variance(test_data, ddof=1)
print(f"Дисперсия со смещением: {test_data_variance_ddof:.2f}")
```

#### 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
first_win = 0.9
second_win = 0.8
third_win = 0.6

shoot_probability = 1/3

shoot_win = shoot_probability * first_win + shoot_probability * second_win + shoot_probability * third_win
print(shoot_win)
```

вероятность того, что выстрел произведен первым спортсменом

```python
first_win_probability = (shoot_probability * first_win)/shoot_win
print(f"{first_win_probability:.2}")
```

вероятность того, что выстрел произведен вторым спортсменом

```python
second_win_probability = (shoot_probability * second_win)/shoot_win
print(f"{second_win_probability:.2}")
```

вероятность того, что выстрел произведен третьим спортсменом

```python
third_win_probability = (shoot_probability * third_win)/shoot_win
print(f"{third_win_probability:.2}")
```

#### 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
good_session_a = .8
good_session_b = .7
good_session_c = .9

good_session = 1/4 * good_session_a + 1/4 * good_session_b + 1/2 * good_session_c
print(good_session)
```

Какова вероятность, что он учится на факультете A

```python
study_a = (1/4 * good_session_a)/good_session
print(f"Вероятность, что студент учится на факультете A: {study_a:.2}")
```

Какова вероятность, что он учится на факультете B

```python
study_b = (1/4 * good_session_b)/good_session
print(f"Вероятность, что студент учится на факультете B: {study_b:.2}")
```

Какова вероятность, что он учится на факультете C

```python
study_c = (1/2 * good_session_c)/good_session
print(f"Вероятность, что студент учится на факультете C: {study_c:.2}")
```

#### 5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
first_bad_detail = .1
first_good_detail = 1 - first_bad_detail
second_bad_detail = .2
second_good_detail = 1 - second_bad_detail
third_bad_detail = .25
third_good_detail = 1 - third_bad_detail
```

вероятность того, что в первый месяц выйдут из строя все детали

```python
probability = first_bad_detail * second_bad_detail * third_bad_detail
print(f"{probability:.2}")
```

вероятность того, что в первый месяц выйдут из строя только две детали

```python
probability = first_good_detail * second_bad_detail * third_bad_detail + first_bad_detail * second_good_detail * third_bad_detail + first_bad_detail * second_bad_detail * third_good_detail
print(f"{probability:.2}")
```

вероятность того, что в первый месяц выйдут из строя хотя бы одна деталь

```python
probability = 1 - first_good_detail * second_good_detail * third_good_detail
print(f"{probability:.2}")
```

вероятность того, что в первый месяц выйдут из строя от одной до двух деталей

```python
probability = 1 - first_good_detail * second_good_detail * third_good_detail - first_bad_detail * second_bad_detail * third_bad_detail
print(f"{probability:.2}")
```
